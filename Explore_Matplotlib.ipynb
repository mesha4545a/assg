```
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
plt.style.use('seaborn-whitegrid')
plt.style.use('seaborn-white')
```

```
sns.get_dataset_names()
```

```
df = sns.load_dataset("tips")

df
```

```
x = df.loc[df['sex'] == 'Male']
y = df.loc[df['sex'] == 'Female']
```

```
plt.figure(figsize=(15,5))
plt.plot(x.tip,'#6666ff', label = 'male tips')  
plt.plot(y.tip,'#FF69B4', label = 'femail tips')
plt.title("male vs femail tips")
plt.axis('tight')
plt.legend();   # specify color by name
 
```

```
plt.figure(figsize=(15,5))
sns.scatterplot(x="tip", y="total_bill",
                hue="sex",
                data=df
                );
 
```

```
plt.figure(figsize=(15,5))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="size", size="size");
```

```
plt.figure(figsize=(15,5))
df["day"].value_counts().plot.pie()
plt.title("Distribution of busy days");
```

```
plt.figure(figsize=(15,5))
df["time"].value_counts().plot.pie()
plt.title("Distribution of busy time");
```

```
plt.figure(figsize=(15,5))
kwargs = dict(histtype='stepfilled', bins=40)
#plt.hist(df.size, **kwargs)
plt.hist(df.tip, **kwargs, label= "tips")
plt.hist(df.total_bill, **kwargs,label= "total bils")
plt.title("total bils and tips")
plt.legend();
```
